10. Как определить, есть ли в данном двумерном массиве пустые столбцы?

````python
def find_empty_columns(arr, empty_value=0):
    if np.isnan(empty_value):
        empty_mask = np.isnan(arr)
    else:
        empty_mask = (arr == empty_value)

    empty_columns = np.all(empty_mask, axis=0)
    return empty_columns

arr1 = np.array([[0, 0, 3],
                 [0, 0, 6],
                 [0, 0, 9]])

arr2 = np.array([[1, np.nan, 3],
                 [4, np.nan, 6]])

print("Пустые столбцы (нули):", find_empty_columns(arr1, 0))
print("Индексы нулевых столбцов в массиве:", np.where(find_empty_columns(arr1, 0))[0])
print("Пустые столбцы (NaN):", find_empty_columns(arr2, np.nan))
print("Индексы нулевых столбцов в массиве:", np.where(find_empty_columns(arr2, np.nan))[0])
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
def get_systolic(bp):
    return int(bp.split('/')[0])

def get_diastolic(bp):
    return int(bp.split('/')[1])

df['Systolic'] = df['Blood Pressure (s/d)'].apply(get_systolic)
df['Diastolic'] = df['Blood Pressure (s/d)'].apply(get_diastolic)

stress_median = df['Stress Levels'].median()
high_stress = df[df['Stress Levels'] > stress_median]
low_stress = df[df['Stress Levels'] <= stress_median]

high_systolic = high_stress['Systolic'].mean()
low_systolic = low_stress['Systolic'].mean()
high_diastolic = high_stress['Diastolic'].mean()
low_diastolic = low_stress['Diastolic'].mean()

systolic_diff_percent = ((high_systolic - low_systolic) / low_systolic) * 100
diastolic_diff_percent = ((high_diastolic - low_diastolic) / low_diastolic) * 100

print(f"Высокий стресс: {high_systolic:.1f}/{high_diastolic:.1f} mmHg")
print(f"Низкий стресс:  {low_systolic:.1f}/{low_diastolic:.1f} mmHg")
print(f"Разница: +{high_systolic - low_systolic:.1f}/+{high_diastolic - low_diastolic:.1f} mmHg")

if abs(systolic_diff_percent) > 5:
    print("Систолическое: разница существенная")
else:
    print("Систолическое: разница незначительная")

if abs(diastolic_diff_percent) > 5:
    print("Диастолическое: разница существенная")
else:
    print("Диастолическое: разница незначительная")
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
bmi_75 = df['BMI'].quantile(0.75)
bmi_30 = df['BMI'].quantile(0.30)
bmi_60 = df['BMI'].quantile(0.60)

healthy_medium = df[(df['BMI'] >= bmi_30) & (df['BMI'] <= bmi_60)]['Chronic Diseases'].isna().sum()
healthy_high = df[df['BMI'] > bmi_75]['Chronic Diseases'].isna().sum()

total_medium = len(df[(df['BMI'] >= bmi_30) & (df['BMI'] <= bmi_60)])
total_high = len(df[df['BMI'] > bmi_75])

percent_healthy_medium = healthy_medium / total_medium * 100
percent_healthy_high = healthy_high / total_high * 100

print(f"Людей без хронических заболеваний:")
print(f"Средний BMI (30-60%): {healthy_medium}/{total_medium} ({percent_healthy_medium:.1f}%)")
print(f"Высокий BMI (>75%): {healthy_high}/{total_high} ({percent_healthy_high:.1f}%)")

if percent_healthy_medium > percent_healthy_high:
    print("У людей с высоким BMI (>75%) хронические заболевания встречаются чаще")
elif percent_healthy_medium < percent_healthy_high:
    print("У людей с высоким BMI (>75%) хронические заболевания встречаются реже")
else:
    print("Частота заболеваний одинаковая")
````

6. Имеет ли связь зрение и способность слышать звуки на высоких частотах? Ответьте на вопрос - какая зависимость между ними и предположите почему зависимость именно такая?

````python
df['Age Group'] = pd.cut(df['Age (years)'], bins=[0, 30, 60, 100], labels=['Молодые', 'Люди среднего возраста', 'Пожилые'])
plt.figure(figsize=(15, 5))

for i, age_group in enumerate(df['Age Group'].unique()):
    plt.subplot(1, 3, i+1)
    group_data = df[df['Age Group'] == age_group]

    plt.scatter(group_data['Vision Sharpness'], group_data['Hearing Ability (dB)'],
                alpha=0.6, label=age_group)

    z = np.polyfit(group_data['Vision Sharpness'].dropna(),
                   group_data['Hearing Ability (dB)'].dropna(), 1)
    p = np.poly1d(z)
    plt.plot(group_data['Vision Sharpness'], p(group_data['Vision Sharpness']), "r--")

    plt.title(f'{age_group}\n(n={len(group_data)})')
    plt.xlabel('Зрение')
    plt.ylabel('Слух (dB)')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
print("\nЗрение и способность слышать звуки на высоких частотах связаны между собой линейной зависимостью, причём с возрастом эта зависимость усиливается.")
print("Зрение и слух ухудшаются вместе не потому, что напрямую зависят друг от друга, а потому что подвержены общим возрастным изменениям организма.")
````
